```python
from collections import defaultdict
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import gzip
import codecs
%matplotlib inline
```

```python
CONGOV_LDA = 'tw_topicos_lda_15.txt.gz'
CONGOV_LSI = 'tw_topicos_lsi_15.txt.gz'

PROGOV_LDA = 'tw_topicos_lda_13.txt.gz'
PROGOV_LSI = 'tw_topicos_lsi_13.txt.gz'

progov = ['globogolpista','dia13diadeluta','domingoeunaovouporque','globoquergolpe','familiamarinhonohsbc']
congov = ['foradilma', 'dilma', 'tchau']
```

```python
def FreqTopicsUsers(filename):
    Freq = defaultdict(int)
    FreqPro = defaultdict(int)
    FreqCon = defaultdict(int)
    f = gzip.open(filename)
    reader = codecs.getreader("utf-8")
    for l in reader(f):
        if len(l.strip())>0 and len(l.split(' >> '))==3:
            user = l.split(' >> ')[0].strip()
            tweet = l.split(' >> ')[2].strip()
            for word in tweet.split():
                if word in progov:
                    FreqPro[user] += 1
                    break
                elif word in congov:
                    FreqCon[user] += 1
                    break
            Freq[user] += 1
    f.close()
    return Freq, FreqPro, FreqCon
```

```python
def PlotT(Freq, Name):
    x,y = zip(*(Freq.items()))
    x = np.array(x)
    y = np.array(y)
    idx = np.argsort(-y)[:10]
    x = x[idx]
    y = y[idx]
    PlotTW(x,y,Name, Name+'_users.eps')
   
def PlotTW( x, y, Name, Fname):
    sns.set(style="white", context="talk")
    rs = np.random.RandomState(7)
    plt.rcParams.update({'legend.fontsize':18})
    plt.figure()
    ax = sns.barplot(np.array(['']*len(x)),y,x, ci=None, palette="Paired", hline=.1)
    plt.title(Name)    
    ax.set_ylabel("Frequency")
    sns.despine(bottom=True)    
    plt.savefig(Fname)
```

```python
Freq, FreqPro, FreqCon = FreqTopicsUsers(PROGOV_LDA)
PlotT(Freq, 'PROGOV')
PlotT(FreqPro, 'PROGOV (Pro)')
PlotT(FreqCon, 'PROGOV (Con)')
```

```python
Freq, FreqPro, FreqCon = FreqTopicsUsers(CONGOV_LDA)
PlotT(Freq, 'CONGOV')
PlotT(FreqPro, 'CONGOV (Pro)')
PlotT(FreqCon, 'CONGOV (Con)')
```
